```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
plt.rcParams['font.size'] = 18
```

```python
df = pd.read_csv('./multi_N25.csv')
df
```

```python
SAVEFIG = True
SAVEPATH = '../Images/Anti_Corruption_Layer/'

C = 1
N = 25

dfPlot = df[df['N_A']+df['N_B']==N]
dfPlot.sort_values(by='N_A', inplace=True)
dfPlot['N_A/N'] = dfPlot['N_A'] / N

plt.plot(dfPlot['N_A/N'], dfPlot['R0'])
plt.fill_between(dfPlot['N_A/N'], dfPlot['R0_low'], dfPlot['R0_up'], alpha=0.3)
plt.xlabel(r'$N_{Slow}$ / N')
plt.ylabel('System Response Time [ms]')
if SAVEFIG:
    plt.tight_layout()
    plt.savefig(SAVEPATH + 'respTime.pdf')
plt.show()

plt.plot(dfPlot['N_A/N'], dfPlot['X0'])
plt.fill_between(dfPlot['N_A/N'], dfPlot['X0_low'], dfPlot['X0_up'], alpha=0.3)
plt.xlabel(r'$N_{Slow}$ / N')
plt.ylabel('System Throughput [r/ms]')
if SAVEFIG:
    plt.tight_layout()
    plt.savefig(SAVEPATH + 'throughput.pdf')
plt.show()

plt.plot(dfPlot['N_A/N'], dfPlot['Uss1']*100, '-', label='SS1')
plt.fill_between(dfPlot['N_A/N'], dfPlot['Uss1_low']*100, dfPlot['Uss1_up']*100, alpha=0.3)
plt.plot(dfPlot['N_A/N'], dfPlot['Uss2']*100, '--', label='SS2')
plt.fill_between(dfPlot['N_A/N'], dfPlot['Uss2_low']*100, dfPlot['Uss2_up']*100, alpha=0.3)
plt.plot(dfPlot['N_A/N'], dfPlot['Uacl']*100, ':', label='ACL')
plt.fill_between(dfPlot['N_A/N'], dfPlot['Uacl_low']*100, dfPlot['Uacl_up']*100, alpha=0.3)
plt.xlabel(r'$N_{Slow}$ / N')
plt.ylabel('Utilization [\%]')
plt.legend()
if SAVEFIG:
    plt.tight_layout()
    plt.savefig(SAVEPATH + 'util.pdf')
plt.show()
```
