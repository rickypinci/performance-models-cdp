```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
plt.rcParams['font.size'] = 18
```

```python
df = pd.read_csv('./multi_N100.csv')
df
```

```python
SAVEFIG = True
SAVEPATH = '../Images/Backends_for_Frontends/'

N = 100

dfPlot = df[df['N_A']+df['N_B']==N]
dfPlot.sort_values(by='N_A', inplace=True)
dfPlot['N_A/N'] = dfPlot['N_A'] / N

plt.plot(dfPlot['N_A/N'], dfPlot['R0'], label='R0')
plt.fill_between(dfPlot['N_A/N'], dfPlot['R0_low'], dfPlot['R0_up'], alpha=0.3)
#plt.plot(dfPlot['N_A/N'], dfPlot['R0_A'], label='R0_A')
#plt.fill_between(dfPlot['N_A/N'], dfPlot['R0_A_low'], dfPlot['R0_A_up'], alpha=0.3)
#plt.plot(dfPlot['N_A/N'], dfPlot['R0_B'], label='R0_B')
#plt.fill_between(dfPlot['N_A/N'], dfPlot['R0_B_low'], dfPlot['R0_B_up'], alpha=0.3)
plt.xlabel(r'$N_{Desktop}$ / N')
plt.ylabel('System Response Time [ms]')
#plt.legend()
if SAVEFIG:
    plt.tight_layout()
    plt.savefig(SAVEPATH + 'respTime.pdf')
plt.show()

plt.plot(dfPlot['N_A/N'], dfPlot['X0'], label='X0')
plt.fill_between(dfPlot['N_A/N'], dfPlot['X0_low'], dfPlot['X0_up'], alpha=0.3)
#plt.plot(dfPlot['N_A/N'], dfPlot['X0_A'], label='X0_A')
#plt.fill_between(dfPlot['N_A/N'], dfPlot['X0_A_low'], dfPlot['X0_A_up'], alpha=0.3)
#plt.plot(dfPlot['N_A/N'], dfPlot['X0_B'], label='X0_B')
#plt.fill_between(dfPlot['N_A/N'], dfPlot['X0_B_low'], dfPlot['X0_B_up'], alpha=0.3)
plt.xlabel(r'$N_{Desktop}$ / N')
plt.ylabel('System Throughput [r/ms]')
if SAVEFIG:
    plt.tight_layout()
    plt.savefig(SAVEPATH + 'throughput.pdf')
#plt.legend()
plt.show()

plt.plot(dfPlot['N_A/N'], dfPlot['Ub1']*100, '-', label='BE1')
plt.fill_between(dfPlot['N_A/N'], dfPlot['Ub1_low']*100, dfPlot['Ub1_up']*100, alpha=0.3)
plt.plot(dfPlot['N_A/N'], dfPlot['Ub2']*100, '--', label='BE2')
plt.fill_between(dfPlot['N_A/N'], dfPlot['Ub2_low']*100, dfPlot['Ub2_up']*100, alpha=0.3)
plt.xlabel(r'$N_{Desktop}$ / N')
plt.ylabel('Utilization [\%]')
plt.legend()
if SAVEFIG:
    plt.tight_layout()
    plt.savefig(SAVEPATH + 'util.pdf')
plt.show()
```
