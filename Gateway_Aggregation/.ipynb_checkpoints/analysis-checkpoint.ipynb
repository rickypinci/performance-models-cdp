```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
plt.rcParams['font.size'] = 18
```

```python
df = pd.read_csv('./multi_N25.csv')
df
```

```python
SAVEFIG = True
SAVEPATH = '../Images/Gateway_Aggregation/'

N = 25

dfPlot = df[df['N_A']+df['N_B']==N]
dfPlot.sort_values(by='N_A', inplace=True)
dfPlot['N_A/N'] = dfPlot['N_A'] / N

plt.plot(dfPlot['N_A/N'], dfPlot['R0'])
plt.fill_between(dfPlot['N_A/N'], dfPlot['R0_low'], dfPlot['R0_up'], alpha=0.3)
plt.xlabel(r'$N_\mathrm{{S3Intensive}}~/~N$')
plt.ylabel('System Response Time [ms]')
if SAVEFIG:
    plt.tight_layout()
    plt.savefig(SAVEPATH + 'respTime.pdf')
plt.show()

plt.plot(dfPlot['N_A/N'], dfPlot['X0'])
plt.fill_between(dfPlot['N_A/N'], dfPlot['X0_low'], dfPlot['X0_up'], alpha=0.3)
plt.xlabel(r'$N_\mathrm{{S3Intensive}}~/~N$')
plt.ylabel('System Throughput [r/ms]')
if SAVEFIG:
    plt.tight_layout()
    plt.savefig(SAVEPATH + 'throughput.pdf')
plt.show()

plt.plot(dfPlot['N_A/N'], dfPlot['Ugw']*100, '-', label='GW')
plt.fill_between(dfPlot['N_A/N'], dfPlot['Ugw_low']*100, dfPlot['Ugw_up']*100, alpha=0.3)
plt.plot(dfPlot['N_A/N'], dfPlot['Us1']*100, '--', label='S1')
plt.fill_between(dfPlot['N_A/N'], dfPlot['Us1_low']*100, dfPlot['Us1_up']*100, alpha=0.3)
plt.plot(dfPlot['N_A/N'], dfPlot['Us2']*100, ':', label='S2')
plt.fill_between(dfPlot['N_A/N'], dfPlot['Us2_low']*100, dfPlot['Us2_up']*100, alpha=0.3)
plt.plot(dfPlot['N_A/N'], dfPlot['Us3']*100, '-.', label='S3')
plt.fill_between(dfPlot['N_A/N'], dfPlot['Us3_low']*100, dfPlot['Us3_up']*100, alpha=0.3)
plt.xlabel(r'$N_\mathrm{{S3Intensive}}~/~N$')
plt.ylabel('Utilization [\%]')
plt.legend(ncol=2)
if SAVEFIG:
    plt.tight_layout()
    plt.savefig(SAVEPATH + 'util.pdf')
plt.show()
```
