```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
plt.rcParams['font.size'] = 18
```

```python
df1 = pd.read_csv('./S_GW_0msec.csv')
df2 = pd.read_csv('./S_GW_50msec.csv')
df3 = pd.read_csv('./S_GW_90msec.csv')
```

```python
N = 100

dfPlot1 = df1[df1['N_A']+df1['N_B']==N]
dfPlot1.sort_values(by='N_A', inplace=True)
dfPlot1['N_A/N'] = dfPlot1['N_A'] / N

plt.plot(dfPlot1['N_A/N'], dfPlot1['R0'])
plt.fill_between(dfPlot1['N_A/N'], dfPlot1['R0_low'], dfPlot1['R0_up'], alpha=0.3)
plt.xlabel('N_A / N')
plt.ylabel('R0 [s]')
plt.show()

plt.plot(dfPlot1['N_A/N'], dfPlot1['X0'])
plt.fill_between(dfPlot1['N_A/N'], dfPlot1['X0_low'], dfPlot1['X0_up'], alpha=0.3)
plt.xlabel('N_A / N')
plt.ylabel('X0 [j/s]')
plt.show()

plt.plot(dfPlot1['N_A/N'], dfPlot1['Us1']*100, label='S1')
plt.fill_between(dfPlot1['N_A/N'], dfPlot1['Us1_low']*100, dfPlot1['Us1_up']*100, alpha=0.3)
plt.plot(dfPlot1['N_A/N'], dfPlot1['Us2']*100, label='S2')
plt.fill_between(dfPlot1['N_A/N'], dfPlot1['Us2_low']*100, dfPlot1['Us2_up']*100, alpha=0.3)
plt.plot(dfPlot1['N_A/N'], dfPlot1['Us3']*100, label='S3')
plt.fill_between(dfPlot1['N_A/N'], dfPlot1['Us3_low']*100, dfPlot1['Us3_up']*100, alpha=0.3)
plt.plot(dfPlot1['N_A/N'], dfPlot1['Ugw']*100, label='GW')
plt.fill_between(dfPlot1['N_A/N'], dfPlot1['Ugw_low']*100, dfPlot1['Ugw_up']*100, alpha=0.3)
plt.xlabel('N_A / N')
plt.ylabel('Util [%]')
plt.legend()
plt.show()
```

```python
N = 100

dfPlot2 = df2[df2['N_A']+df2['N_B']==N]
dfPlot2.sort_values(by='N_A', inplace=True)
dfPlot2['N_A/N'] = dfPlot2['N_A'] / N

plt.plot(dfPlot2['N_A/N'], dfPlot2['R0'])
plt.fill_between(dfPlot2['N_A/N'], dfPlot2['R0_low'], dfPlot2['R0_up'], alpha=0.3)
plt.xlabel('N_A / N')
plt.ylabel('R0 [s]')
plt.show()

plt.plot(dfPlot2['N_A/N'], dfPlot2['X0'])
plt.fill_between(dfPlot2['N_A/N'], dfPlot2['X0_low'], dfPlot2['X0_up'], alpha=0.3)
plt.xlabel('N_A / N')
plt.ylabel('X0 [j/s]')
plt.show()

plt.plot(dfPlot2['N_A/N'], dfPlot2['Us1']*100, label='S1')
plt.fill_between(dfPlot2['N_A/N'], dfPlot2['Us1_low']*100, dfPlot2['Us1_up']*100, alpha=0.3)
plt.plot(dfPlot2['N_A/N'], dfPlot2['Us2']*100, label='S2')
plt.fill_between(dfPlot2['N_A/N'], dfPlot2['Us2_low']*100, dfPlot2['Us2_up']*100, alpha=0.3)
plt.plot(dfPlot2['N_A/N'], dfPlot2['Us3']*100, label='S3')
plt.fill_between(dfPlot2['N_A/N'], dfPlot2['Us3_low']*100, dfPlot2['Us3_up']*100, alpha=0.3)
plt.plot(dfPlot2['N_A/N'], dfPlot2['Ugw']*100, label='GW')
plt.fill_between(dfPlot2['N_A/N'], dfPlot2['Ugw_low']*100, dfPlot2['Ugw_up']*100, alpha=0.3)
plt.xlabel('N_A / N')
plt.ylabel('Util [%]')
plt.legend()
plt.show()
```

```python
SAVEFIG = True
SAVEPATH = '../Images/Gateway_Offloading/'

plt.plot(dfPlot1['N_A/N'], dfPlot1['R0'], label=r'$S_{Gateway}$ = 0 ms')
plt.fill_between(dfPlot1['N_A/N'], dfPlot1['R0_low'], dfPlot1['R0_up'], alpha=0.3)
plt.plot(dfPlot2['N_A/N'], dfPlot2['R0'], label=r'$S_{Gateway}$ = 50 ms')
plt.fill_between(dfPlot2['N_A/N'], dfPlot2['R0_low'], dfPlot2['R0_up'], alpha=0.3)
plt.plot(dfPlot3['N_A/N'], dfPlot3['R0'], label=r'$S_{Gateway}$ = 90 ms')
plt.fill_between(dfPlot3['N_A/N'], dfPlot3['R0_low'], dfPlot3['R0_up'], alpha=0.3)
plt.xlabel(r'$N_A$ / N')
plt.ylabel('System Latency [s]')
plt.legend()
if SAVEFIG:
    plt.tight_layout()
    plt.savefig(SAVEPATH + 'respTime.pdf')
plt.show()
```
