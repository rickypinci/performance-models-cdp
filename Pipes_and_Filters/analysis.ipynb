```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
plt.rcParams['font.size'] = 18
```

```python
dfJ1 = pd.read_csv('./joint/Core1_N50.csv')
dfJ2 = pd.read_csv('./joint/Core2_N50.csv')
dfS = pd.read_csv('./separated/Core1_N50.csv')
```

```python
N = 50

dfJ1 = dfJ1[dfJ1['N_A']+dfJ1['N_B']==N]
dfJ1.sort_values(by='N_A', inplace=True)
dfJ1['N_A/N'] = dfJ1['N_A'] / N

dfJ2 = dfJ2[dfJ2['N_A']+dfJ2['N_B']==N]
dfJ2.sort_values(by='N_A', inplace=True)
dfJ2['N_A/N'] = dfJ2['N_A'] / N

dfS = dfS[dfS['N_A']+dfS['N_B']==N]
dfS.sort_values(by='N_A', inplace=True)
dfS['N_A/N'] = dfS['N_A'] / N
```

```python
SAVEFIG = True
SAVEPATH = '../Images/Pipes_and_Filters/'

plt.plot(dfJ1['N_A/N'], dfJ1['R0'], '-', label='Joint (1)')
plt.fill_between(dfJ1['N_A/N'], dfJ1['R0_low'], dfJ1['R0_up'], alpha=0.3)
plt.plot(dfJ2['N_A/N'], dfJ2['R0'], '--', label='Joint (2)')
plt.fill_between(dfJ2['N_A/N'], dfJ2['R0_low'], dfJ2['R0_up'], alpha=0.3)
plt.plot(dfS['N_A/N'], dfS['R0'], ':', label='Separated')
plt.fill_between(dfS['N_A/N'], dfS['R0_low'], dfS['R0_up'], alpha=0.3)
plt.xlabel(r'$N_{Freemium}$ / N')
plt.ylabel('System Response Time [ms]')
plt.yticks(range(300,701,100))
plt.legend(fontsize=17)
if SAVEFIG:
    plt.tight_layout()
    plt.savefig(SAVEPATH + 'respTime.pdf')
plt.show()

plt.plot(dfJ1['N_A/N'], dfJ1['X0'], '-', label='Joint (x1)')
plt.fill_between(dfJ1['N_A/N'], dfJ1['X0_low'], dfJ1['X0_up'], alpha=0.3)
plt.plot(dfJ2['N_A/N'], dfJ2['X0'], '--', label='Joint (x2)')
plt.fill_between(dfJ2['N_A/N'], dfJ2['X0_low'], dfJ2['X0_up'], alpha=0.3)
plt.plot(dfS['N_A/N'], dfS['X0'], ':', label='Separated')
plt.fill_between(dfS['N_A/N'], dfS['X0_low'], dfS['X0_up'], alpha=0.3)
plt.xlabel(r'$N_{Freemium}$ / N')
plt.ylabel('System Throughput [r/ms]')
if SAVEFIG:
    plt.tight_layout()
    plt.savefig(SAVEPATH + 'throughput.pdf')
#plt.legend()
plt.show()
```
