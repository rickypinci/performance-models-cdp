```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
plt.rcParams['font.size'] = 18
```

```python
ERR = 10**-9

N = 50
S1A = 14
S2A = 8
S1B = 6
S2B = 15

df1 = pd.read_csv('./multi_N50.csv')

df1 = df1[(df1['N_A']+df1['N_B']==N) &
          (np.abs(df1['r_A_APP'] - 1/S1A) < ERR) &
          (np.abs(df1['r_A_STO'] - 1/S2A) < ERR) &
          (np.abs(df1['r_B_APP'] - 1/S1B) < ERR) &
          (np.abs(df1['r_B_STO'] - 1/S2B) < ERR)]
df1.sort_values(by='N_A', inplace=True)
df1['N_A/N'] = df1['N_A'] / N
```

```python
SAVEFIG = True
SAVEPATH = '../Images/Static_Content_Hosting/'

plt.plot(df1['N_A/N'], df1['R0'])
plt.fill_between(df1['N_A/N'], df1['R0_low'], df1['R0_up'], alpha=0.3)
plt.xlabel(r'$N_{CompIntensive}$ / N')
plt.ylabel('System Response Time [ms]')
plt.yticks(range(550,800,50))
if SAVEFIG:
    plt.tight_layout()
    plt.savefig(SAVEPATH + 'respTime.pdf')
plt.show()

plt.plot(df1['N_A/N'], df1['X0'])
plt.fill_between(df1['N_A/N'], df1['X0_low'], df1['X0_up'], alpha=0.3)
plt.xlabel(r'$N_{CompIntensive}$ / N')
plt.ylabel('System Throughput [r/ms]')
if SAVEFIG:
    plt.tight_layout()
    plt.savefig(SAVEPATH + 'throughput.pdf')
plt.show()

plt.plot(df1['N_A/N'], df1['Uapp']*100, '-', label='Computation')
plt.fill_between(df1['N_A/N'], df1['Uapp_low']*100, df1['Uapp_up']*100, alpha=0.3)
plt.plot(df1['N_A/N'], df1['Usto']*100, '--', label='Storage')
plt.fill_between(df1['N_A/N'], df1['Usto_low']*100, df1['Usto_up']*100, alpha=0.3)
plt.xlabel(r'$N_{CompIntensive}$ / N')
plt.ylabel('Utilization [\%]')
plt.legend(loc='lower center')
if SAVEFIG:
    plt.tight_layout()
    plt.savefig(SAVEPATH + 'util.pdf')
plt.show()
```
